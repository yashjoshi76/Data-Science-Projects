```python
#import the required modules
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#inspect the wiki page and find the class/id name
wiki_url = 'https://en.wikipedia.org/wiki/List_of_breweries_in_Ireland'
table_class = 'wikitable'
#make a request and store the response
response = requests.get(wiki_url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
#use find_all(class_ = 'className')
brews_table= soup.find_all(class_ = table_class)
```

```python
brews_table
```

```python
#store the table content in a dataframe
df = pd.read_html(str(brews_table))
```

```python
df
```

```python
#since the df object has multiple data frames we use a for loop
for i in df:
    i.to_csv("beer_brews.csv", index = False)
```
